```python
def calculate_demographic_data(print_data=True):
    # Read data from file
    df = pd.read_csv("adult.data.csv", delimiter=",")  # Se abre el archivo 
    print(df.head())  # Muestra las primeras filas
    
    # How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
    race_count = None

    # What is the average age of men?
    average_age_men = None

    # What is the percentage of people who have a Bachelor's degree?
    percentage_bachelors = None

    # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
    # What percentage of people without advanced education make more than 50K?

    # with and without `Bachelors`, `Masters`, or `Doctorate`
    higher_education = None
    lower_education = None

    # percentage with salary >50K
    higher_education_rich = None
    lower_education_rich = None

    # What is the minimum number of hours a person works per week (hours-per-week feature)?
    min_work_hours = None

    # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
    num_min_workers = None

    rich_percentage = None

    # What country has the highest percentage of people that earn >50K?
    highest_earning_country = None
    highest_earning_country_percentage = None

    # Identify the most popular occupation for those who earn >50K in India.
    top_IN_occupation = None

    # DO NOT MODIFY BELOW THIS LINE

    if print_data:
        print("Number of each race:\n", race_count) 
        print("Average age of men:", average_age_men)
        print(f"Percentage with Bachelors degrees: {percentage_bachelors}%")
        print(f"Percentage with higher education that earn >50K: {higher_education_rich}%")
        print(f"Percentage without higher education that earn >50K: {lower_education_rich}%")
        print(f"Min work time: {min_work_hours} hours/week")
        print(f"Percentage of rich among those who work fewest hours: {rich_percentage}%")
        print("Country with highest percentage of rich:", highest_earning_country)
        print(f"Highest percentage of rich people in country: {highest_earning_country_percentage}%")
        print("Top occupations in India:", top_IN_occupation)

    return {
        'race_count': race_count,
        'average_age_men': average_age_men,
        'percentage_bachelors': percentage_bachelors,
        'higher_education_rich': higher_education_rich,
        'lower_education_rich': lower_education_rich,
        'min_work_hours': min_work_hours,
        'rich_percentage': rich_percentage,
        'highest_earning_country': highest_earning_country,
        'highest_earning_country_percentage':
        highest_earning_country_percentage,
        'top_IN_occupation': top_IN_occupation
    }




    

calculate_demographic_data()
```

```python
############### Total de personas por país y por ingresos  ################################## 

# Contar personas por país
people_per_country = df["native-country"].value_counts().sort_index().reset_index() # Reset_index convierte el índice en native-country en una columna normal. Para que la tabla quede organizada como un DataFrame normal y no como una Serie con índice jerárquico.
people_per_country.columns = ["Country", "Count"]
display(people_per_country)

# Contar personas por país y salario (>50K y <=50K)
income_percentage = df.groupby(["native-country", "salary"]).size().unstack(fill_value=0).reset_index() # Convierte los valores de la columna "salary" en columnas separadas. fill_value=0 asegura que si un país no tiene alguna categoría de salario, el valor sea 0 en lugar de NaN.

# Renombrar columnas (asegúrate de que existen ambas categorías de salario)
if ">50K" in income_percentage.columns and "<=50K" in income_percentage.columns:
    income_percentage.columns = ["Country", "Count <=50K", "Count >50K"]
elif ">50K" in income_percentage.columns:
    income_percentage.columns = ["Country", "Count >50K"]
elif "<=50K" in income_percentage.columns:
    income_percentage.columns = ["Country", "Count <=50K"]

# Convertir a DataFrame y mostrar
income_percentage_df = pd.DataFrame(income_percentage)
display(income_percentage_df)
```

```python
# Total de personas por país

total_by_country = df["native-country"].value_counts().reset_index()
total_by_country.columns = ["Country", "Total"]

# Total de personas que ganan más de 50K por país
high_income_by_country = df[df["salary"] == ">50K"]["native-country"].value_counts().reset_index()
high_income_by_country.columns = ["Country", "High Income Count"]

# Unir ambas tablas
income_distribution = total_by_country.merge(high_income_by_country, on="Country", how="left").fillna(0)

# Calcular porcentaje de personas que ganan más de 50K en cada país
income_distribution["High Income (%)"] = round(
    (income_distribution["High Income Count"] / income_distribution["Total"]) * 100, 2
)

# Asegurar que la columna sea numérica antes de ordenar
income_distribution["High Income (%)"] = pd.to_numeric(income_distribution["High Income (%)"], errors="coerce")

# Ordenar de menor a mayor por porcentaje de ingresos >50K
income_distribution_sorted = income_distribution.sort_values(by="High Income (%)", ascending=True)

# Mostrar resultado ordenado
display(income_distribution_sorted)  # Si usas Jupyter Notebook
```

```python
# Redondear el porcentaje a un solo decimal
income_distribution["High Income (%)"] = income_distribution["High Income (%)"].round(1)

# Encontrar el país con el mayor porcentaje de ingresos >50K
max_income_country = income_distribution.loc[income_distribution["High Income (%)"].idxmax()]

# Mostrar el resultado con un solo decimal
print(f"El país con la mayor proporción de ingresos >50K es {max_income_country['Country']} con {max_income_country['High Income (%)']}%")
```
